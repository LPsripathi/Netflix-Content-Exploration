Staring with Netflix data, there are around 8807 records containing info of Movies and TV shows on Netflix Platform. So to conduct EDA and Visuals, Lets start with important step of cleaning Data.



```python
import pandas as pd
```

## ***Objectives set for Cleaning data -***



---




1.   Removing all the missing values from the data
2.   Every show_id is identified as an object which is rectified by removing letter 's' and converting it into int
3.   Convert data_added to date_time format
4.   Seperate Movie and TV Show data into 2 seperate data frames.

```python
data_df.isnull().sum()
```

```python
data_df.dropna(inplace=True)
```

```python
data_df.isnull().sum()
```

```python
data_df.head()
```

```python
data_df['show_id']=data_df['show_id'].str.replace('s','')
```

```python
data_df
```

```python
data_df['show_id']=data_df['show_id'].astype(int)
```

```python
data_df.info()
```

```python
data_df['date_added'] = pd.to_datetime(data_df['date_added'].str.strip(), format='%B %d, %Y')
```

```python
data_df.info()
```

```python
data_df.head()
```

```python
tv_df = data_df[data_df['type'] == 'TV Show']
```

```python
movie_df = data_df[data_df['type'] == 'Movie']
```

```python
tv_df.shape
```

```python
movie_df.shape
```

# ***Exploratory Data Analysis***


---

Performing Data Analysis on cleaned data. We have combined dataset of Movies and TV shows.

## ***EDA 1- Comparing movies and TV show content on Netflix platform.***

```python
import matplotlib.pyplot as plt
```

```python
data_df['type'].value_counts().plot(kind='pie',autopct='%1.1f%%', colors=['tab:blue', 'skyblue'])
plt.title("Streaming Content- Movies vs TV Shows")
plt.ylabel('')
plt.show()
```

```python
data_df['type'].value_counts().plot(kind='bar')
plt.title("Streaming Content- Movies vs TV Shows")
plt.xticks(rotation=45)
plt.xlabel('Content type')
plt.ylabel('Number of shows')
plt.show()
```

TV shows are very less compared to Movies streaming on Netflix.

## ***EDA 2- Top 10 countries with highest content streaming***

```python
countries_df=data_df['country'].str.split(',').explode().str.split()
```

```python
countries_df.head()
```

```python
top_countries=countries_df.value_counts().head(10)
```

```python
top_countries.plot(kind='bar')
plt.xlabel('Countries')
plt.ylabel('No of shows streamed on')
plt.title('Top 10 countries with most content')
plt.xticks(rotation=45)
plt.show()
```

USA and India looks to be having the highest Streaming content on Netflix Platform.

## ***EDA 3- Content Added Over time***

```python
data_df['year_added']=data_df['date_added'].dt.year
```

```python
data_df.columns
```

```python
data_df['year_added'].value_counts()
```

```python
data_df['year_added'].value_counts().sort_index().plot(kind='line',marker='o')
plt.xlabel('Year added on Netflix')
plt.ylabel('Shows')
plt.title('Content added over Years')
plt.show()
```

2019 is the year where lots of shows and movies were added onto the Netflix platform.
